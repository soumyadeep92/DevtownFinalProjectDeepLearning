## 5. Data filtering and analysis

```python
Start_Date=''
End_Date=''
act_year=[]
date_join=df['Dateofjoining']
date_end=df['LastWorkingDate']
for i in range(len(date_join)):
  start=[]
  end=[]
  Start_Date=date_join[i]
  start.append(Start_Date.split('-'))
  start_year=int(start[0][0])
  start_months=int(start[0][1])
  start_days=int(start[0][2])
  End_Date=date_end[i]
  end.append(End_Date.split('-'))
  end_year=int(end[0][0])
  end_months=int(end[0][1])
  end_days=int(end[0][2])
  act_year.append(float(end_year-start_year)+float((end_months-start_months)/12)+float((end_days-start_days)/365))
print(act_year)
```

```python
print(df[0:10000:10]['Age'].unique().shape)
print(df[0:10000:10]['Age'].unique())
print(df[0:10000:10]['Education_Level'].unique().shape)
print(df[0:10000:10]['Education_Level'].unique())
print(df[0:10000:10]['Designation'].unique().shape)
print(df[0:10000:10]['Designation'].unique())
print(df[0:10000:10]['Total Business Value'].unique().shape)
```

```python
data=pd.DataFrame()
data['Age']=df[0:10000:10]['Age']
data['Education_Level']=df[0:10000:10]['Education_Level']
data['Experience']=act_year[0:10000:10]
data['Designation']=df[0:10000:10]['Designation']
data['Total_Business_Value']=df[0:10000:10]['Total Business Value']
print(data.head())
```

```python
feature_ = ['Experience','Total_Business_Value','Salary']
def get_percentile(feature, q_range, dat):
  dist = dat[feature].describe()[str(q_range) + '%']
  return round(dist,2)
```

```python
def render_counterplot(feat, dat):
    fig=plt.figure(figsize=(18, 20))
    fig.add_subplot(4, 3, 1 + 1)
        
    q1 = get_percentile(feat, 25, dat)
    q2 = get_percentile(feat, 50, dat)
    q3 = get_percentile(feat, 75, dat)
          
    sns.histplot(data=dat, x=feat, kde=True, color = 'orange')
        
    plt.axvline(q1, linestyle='--', color='green', label='Q1')
    plt.axvline(q2, color='red', label='Q2 (Median)')
    plt.axvline(q3, linestyle='--',  color='black', label='Q3')
    plt.legend()
    plt.show()
    
render_counterplot(feature_[0],data)
render_counterplot(feature_[1],data)
```

## 8. Processing Final data values for prediction

```python
p1=[]
p2=[]
for i in data[0:1000:75]['Experience']:
  p1.append(i)
for j in y[0:1000:75]['Salary']:
  p2.append(j)
plt.bar(p1,p2,width=0.1)
plt.title('Plot of Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.savefig('Plot 2')
plt.show()
```

```python
print(data.isnull().sum())
print(y.isnull().sum())
```

```python
categorical_var='Education_Level'

cat_list = 'var'+'_'+categorical_var
cat_list = pd.get_dummies(data[categorical_var], prefix=categorical_var)
data1 = data.join(cat_list)
data = data1
```

```python
data_vars = data.columns.values.tolist()
print(data_vars)
```

```python
to_keep = [i for i in data_vars if i not in categorical_var]
to_keep
```

```python
data_final = data[to_keep]
data_final.columns.values
print(data_final)
```

```python
sns.set(rc={'figure.figsize':(11, 8)})
sns.distplot(y['Salary'], bins= 30)
plt.savefig('Plot 3')
plt.show()
```

## 9. Plotting regression line

```python
plt.scatter(data_final[0:]['Experience'],y)
plt.scatter(data_final[0:]['Age'],y)
m, b=np.polyfit(data_final[0:]['Experience'],y,1)
m1, b1=np.polyfit(data_final[0:]['Age'],y,1)
plt.plot(data_final[0:]['Experience'],m*data_final[0:]['Experience']+b,color='red')
plt.plot(data_final[0:]['Age'],m1*data_final[0:]['Age']+b1,color='red')
plt.title('Plot of Experience & Age vs Salary of given dataset')
plt.ylabel('Salary')
plt.legend(labels=('Experience','Age'))
plt.savefig('regr1')
plt.show()
```

## 10. Test and Train datasets created

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data_final,y,test_size=0.4,random_state=42)
```

## 11. Linear Regression 

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred_data=model.predict(data_final)
# print(y_pred_data)
```

## 12. Dimensionally reduced plot for salary prediction

```python
plt.scatter(data_final[0:]['Experience'],y_pred_data)
plt.scatter(data_final[0:]['Age'],y_pred_data)
plt.title('Dimensionally Reduced 1D plot for Experience & Age vs Salary')
plt.ylabel('Predicted Salaries for all independent variables')
plt.legend(labels=('Experience','Age'))
plt.savefig('regr2')
plt.show()
```

```python
plt.scatter(data_final[0:]['Total_Business_Value'],y_pred_data)
plt.title('Dimensionally Reduced 1D plot for Business Values vs Salary')
plt.xlabel('Total Business Values')
plt.ylabel('Predicted Salaries for all independent variables')
plt.savefig('regr3')
plt.show()
```

## 13. Post regression analysis

```python
print("The equation of the multivariate linear regression is: ")
col_vars=list(data_final.columns.values)
print(col_vars)
```

```python
x_eq=''
for i in range(len(col_vars)):
  x_eq+=str(model.coef_[0][i])+"*"+col_vars[i]+"+"
x_eq=x_eq+str(int(model.intercept_))
print(x_eq)
```

## 14. Prediction of output

```python
y_pred=model.predict(x_test)
print("Two Sampled Predicted salaries on the test set x_test is: ")
print(int(y_pred[0]))
print(int(y_pred[1]))
```

```python
print("Estimated Salary for Rajiv, age 22 and having 3 years of experience is: ", model.predict([[22, 3, 3, 200000, 0, 0, 1]])[0][0].round(2))
```

## 15. Calculating the metrics of result

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
score = r2_score(y_test, y_pred)
print("r2 score is ", score)
absolute_error = mean_absolute_error(y_test, y_pred)
mean_error = mean_squared_error(y_test, y_pred)
print("Mean absolute error is ", absolute_error)
print("Mean squared error is ", mean_error)
```

## 16. Increase accuracy

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10,random_state=0)
regressor.fit(data_final, y)
```

## 17. Plot of Predicted Salaries of Higher Accuracy  

```python
y_pred_rand=regressor.predict(data_final)
```

```python
plt.scatter(data_final[0:]['Experience'],y_pred_rand)
plt.scatter(data_final[0:]['Age'],y_pred_rand)
plt.title('Dimensionally Reduced 1D plot for Experience & Age vs Salary')
plt.ylabel('Predicted Salaries for all independent variables')
plt.legend(labels=('Experience','Age'))
plt.savefig('regrn')
plt.show()
```

## 18. Plot Importance metrics

```python
model_ranks = pd.Series(regressor.feature_importances_, index = x_train.columns, name='Importance')
ax = model_ranks.plot(kind = 'barh')
```

```python
print("Accurately estimated Salary for Rajiv, age 22 and having 3 years of experience is: ", regressor.predict([[22, 3, 3, 200000, 0, 0, 1]])[0].round(2))
```

## 19. Final metrics Calculation

```python
y_pred_random = regressor.predict(x_test)
score = r2_score(y_test, y_pred_random)
print("r2 score is ", score)
absolute_error = mean_absolute_error(y_test, y_pred_random)
mean_error = mean_squared_error(y_test, y_pred_random)
print("Mean absolute error is ", absolute_error)
print("Mean squared error is ", mean_error)
```
